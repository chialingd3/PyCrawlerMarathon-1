```python
import requests 

r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text
response
```

```python
import json
data = json.loads(response)

for d in data:
    print(d['title'])
    print(d['createdAt'])
    print(d['commentCount'])
    print(d['likeCount'])
```
