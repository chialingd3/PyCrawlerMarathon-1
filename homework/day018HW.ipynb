```python
import requests
url = 'https://www.zhihu.com/explore'
r = requests.get(url)
r.encoding = 'utf-8'
print(r.text)
```

```python
import requests
url = 'http://www.zhihu.com'

headers = {
    'Referer': 'https://www.zhihu.com/explore'
    'Sec-Fetch-Dest': 'style'
}

r = requests.get(url, headers=headers)

r.encoding = 'utf-8'
print(r.text[0:600])
```
